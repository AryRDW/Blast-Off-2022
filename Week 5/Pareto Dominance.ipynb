### Importing the necessary libraries

```python
import pygmo as pg
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading data

```python
df = pd.read_csv('rocket_data.csv')
mass = df['MASS']
cost = df['COST']
```

### Plotting data

```python
plt.figure(figsize=(12, 8))
plt.title("Rocket Data")
plt.xlabel("Mass")
plt.ylabel("Cost")
plt.scatter(mass, cost)
plt.grid()
plt.show()
```

### Finding Pareto Dominance

```python
i = 0
while (len(mass) > 0):
    points = [[-1*mass[i], cost[i]] for i in range(len(cost))]
    di = pg.non_dominated_front_2d(points=points)
    plt.figure(figsize=(12, 8))
    plt.scatter(mass, cost)
    plt.scatter([mass[j] for j in di], [cost[j] for j in di])
    plt.grid()
    plt.title("Dominance = " + str(i))
    plt.xlabel("Mass")
    plt.ylabel("Cost")
    plt.show()
    new_mass = []
    new_cost = []
    for j in range(len(mass)):
        if j not in di:
            new_mass.append(mass[j])
            new_cost.append(cost[j])
    mass = new_mass
    cost = new_cost
    i += 1
```
